```python
import pandas as pd
import numpy as np
import plotly.express as px
from matplotlib import pyplot as plt
import seaborn as sns

pd.options.display.float_format = '{:,.2f}'.format
pd.options.display.max_rows = None
pd.options.display.max_columns = None

sns.set()
```

```python
data = pd.read_csv('houses_Madrid.csv')
```

```python
data.info()
```

```python
data = data.dropna(axis=1, how='all') #remove column(axis=1), where all values are none(how='all')
data.info()
```

```python
data.duplicated().sum()
```

```python
df = data[['id', 'subtitle', 'sq_mt_built', 'sq_mt_useful', 'n_rooms', 'n_bathrooms', 'n_floors', 'sq_mt_allotment', 'rent_price', 'buy_price', 'buy_price_by_area','built_year', 'parking_price']]
df.info()
```

```python
df=df.rename(columns={'subtitle' : 'city'})
```

```python
df = df[['id', 'city', 'sq_mt_built', 'n_rooms', 'n_bathrooms', 'rent_price', 'buy_price', 'buy_price_by_area']]
df.info()
```

```python
df.n_rooms.describe()
```

```python
df.drop(df.index[df.n_rooms == 0], axis=0, inplace=True)
```

```python
df.n_bathrooms.mode()
```

```python
df.n_bathrooms.fillna(1, inplace=True)
df.n_bathrooms = df.n_bathrooms.replace(0,1)
```

```python
df.info()
```

```python
df.sample(10)
```

```python
df.n_bathrooms = df.n_bathrooms.astype(int)
```

```python
df.info()
```

```python
#drop sq_mt_built NaN rows
df.dropna(subset='sq_mt_built', inplace=True)
```

```python
df.info()
```

```python
df.sample(10)
```

```python
#change negative values in Rent price column to Zero
for value in df.rent_price.values:
    if value < 0:
        df.rent_price = 0
```

```python
df[df.rent_price < 0]
```

```python
#split City Column to (district, city)
df['district'] = df.city.str.split(',', expand=True)[0]
```

```python
df.head()
```

```python
df = df[['id', 'district', 'sq_mt_built', 'n_rooms', 'n_bathrooms', 'rent_price',
       'buy_price', 'buy_price_by_area']]
df.head()
```

```python
data.head()
```

```python
data[['raw_address','is_exact_address_hidden','street_name']].info()
```

```python
data['operation'].unique()
```

```python
# Remove rent_price
df.drop(columns=['rent_price'], inplace=True)
```
